```python
#Creating a function that categorize the NatureTitle
def classify_nature_type(title: str) -> str:
    t = str(title).lower().strip()
    
    # The order of checks is crucial for correctly handling overlapping terms (e.g., burns, fractures).

    # -----------------------------------------------------------
    # 1. HIGH SEVERITY / SPECIFIC TRAUMA
    # -----------------------------------------------------------

    # ---------- Gunshot Wounds ----------
    if "gunshot wound" in t:
        return "Gunshot Wounds"

    # ---------- Amputations/Avulsions/Enucleations ----------
    elif (
        "amput" in t or
        "avulsion" in t or
        "enucleation" in t
    ):
        return "Amputations"
    
    # ---------- Poisoning/Toxic/Allergenic Effects ----------
    # Includes asphyxiation/allergic reactions related to exposure.
    elif (
        "poison" in t or
        "toxic" in t or
        "noxious" in t or
        "allergenic" in t or
        "anaphylactic" in t or
        "anaphylaxis" in t or
        "asphyxia" in t and ("poison" in t or "toxic" in t or "allergenic" in t)
    ):
        return "Poisoning/Toxic Effects"

    # ---------- Electrocution/Electrical Injuries ----------
    # Must precede BURNS to capture all degrees of electrical burns/shocks.
    elif (
        "electrocution" in t or
        "electric shock" in t or
        "electrical burn" in t or
        "electrical injuries" in t or
        "electrical current" in t
    ):
        return "Electrocution"

    # ---------- CHEMICAL EXPOSURE/BURNS/CORROSIONS ----------
    # Must precede BURNS.
    elif (
        "chemical burn" in t or
        "corrosion" in t
    ):
        return "Chemical Exposure"

    # ---------- FRACTURES ----------
    # Includes terms combining fractures with other injuries (except for highly specific joint trauma).
    elif "fracture" in t:
        return "Fractures"

    # ---------- CRUSHING INJURIES ----------
    elif "crushing" in t:
        return "Crushing Injuries"

    # ---------- DISLOCATIONS ----------
    # Includes terms combining dislocations with other injuries.
    elif "dislocation" in t:
        return "Dislocations"


    # -----------------------------------------------------------
    # 2. INTERNAL/SEVERE TRAUMA & CNS INJURIES
    # -----------------------------------------------------------
    
    # ---------- INTERNAL TRAUMA ----------
    # Injuries to internal organs, blood vessels, and intracranial structures.
    elif (
        "intracranial injur" in t or
        "spinal cord injur" in t or
        "cerebral and other intracranial hemorrhages" in t or
        "concussion" in t or
        "closed trauma involving internal organs" in t or
        "internal injuries to organs and blood vessels of the trunk" in t or
        "injuries to the brain, spinal cord" in t or
        "internal injur" in t or
        "severe wounds" in t
    ):
        return "Internal/Severe Trauma"


    # -----------------------------------------------------------
    # 3. BURNS (Thermal)
    # -----------------------------------------------------------

    # ---------- THERMAL BURNS ----------
    # General heat burns, after filtering electrical and chemical.
    elif (
        "heat (thermal) burn" in t or
        "thermal burn" in t or
        "other burn" in t or
        "burn" in t # Catch-all for burns not already classified
    ):
        return "Thermal Burns"

    # -----------------------------------------------------------
    # 4. MUSCULOSKELETAL & SOFT TISSUE
    # -----------------------------------------------------------

    # ---------- TEARS/SPRAINS/STRAINS/DISC DISORDERS ----------
    # Specific injuries to soft tissue structures.
    elif (
        "tear" in t or
        "strain" in t or
        "sprain" in t or
        "meniscus" in t or
        "cartilage fracture" in t or
        "disc disorders" in t or
        "herniated disc" in t or
        "whiplash" in t
    ):
        return "Sprains/Tears/Strains"

    # ---------- OPEN WOUNDS/CUTS ----------
    # Specific surface/flesh wounds.
    elif (
        "cut" in t or
        "laceration" in t or
        "puncture" in t or
        "open wound" in t or
        "abrasions" in t or
        "scratches" in t or
        "blisters" in t or
        "eye abrasion" in t
    ):
        return "Open Wounds/Cuts"

    # ---------- SOFT TISSUE/BRUISES/CONTUSIONS ----------
    # Broader surface wounds/injuries.
    elif (
        "soft tissue" in t or
        "bruise" in t or
        "contusion" in t or
        "bursitis" in t or
        "surface, flesh wounds" in t
    ):
        return "Soft Tissue"

    # -----------------------------------------------------------
    # 5. ILLNESS, INFECTIONS, ENVIRONMENTAL & OTHER MEDICAL
    # -----------------------------------------------------------
    
    # ---------- INFECTIOUS/IMMUNITY DISORDERS ----------
    elif (
        "infectious" in t or
        "infection" in t or
        "viral" in t or
        "bacterial" in t or
        "mycoses" in t or
        "rabies" in t or
        "rickettsioses" in t or
        "legionnaires" in t or
        "leptospirosis" in t or
        "mrsa" in t or
        "influenza" in t or
        "immunity disorders" in t or
        "cellulitis and abscess" in t or
        "exposure to disease" in t
    ):
        return "Infectious"

    # ---------- RESPIRATORY SYSTEM ----------
    elif (
        "respiratory" in t or
        "pneumonia" in t or
        "pneumonitis" in t or
        "pneumoconioses" in t or
        "asthma" in t or
        "chronic obstructive pulmonary" in t or
        "pulmonary edema" in t or
        "pulmonary embolism" in t
    ):
        return "Respiratory"

    # ---------- CIRCULATORY/CARDIOVASCULAR SYSTEM (Non-Traumatic) ----------
    elif (
        "circulatory system" in t or
        "cardiovascular system" in t or
        "heart failure" in t or
        "phlebitis" in t or
        "thrombosis" in t or
        "hypertensive disease" in t or
        "ischemic heart disease" in t or
        "angina" in t or
        "tachycardia" in t or
        "embolism" in t or
        "myocardial infarction" in t or
        "transient ischemia attacks" in t or
        "stroke" in t
    ):
        return "Circulatory/Cardiovascular"
    
    # ---------- ENDOCRINE/METABOLIC/BLOOD/HERNIAS ----------
    # Other systemic non-traumatic conditions.
    elif (
        "endocrine" in t or
        "metabolic" in t or
        "anemia" in t or
        "hern" in t
    ):
        return "Illness/Disorders"

    # ---------- ENVIRONMENTAL EFFECTS (Heat/Cold/Pressure/Drowning) ----------
    elif (
        "heat stroke" in t or
        "heat exhaustion" in t or
        "heat syncope" in t or
        "frostbite" in t or
        "hypothermia" in t or
        "caisson disease" in t or
        "drowning" in t or
        "effects of heat" in t or
        "effects of reduced temperature" in t or
        "asphyxia" in t # Non-toxic asphyxia
    ):
        return "Environmental Effects"

    # ---------- INFLAMMATION/SKIN REACTIONS ----------
    elif (
        "inflammation" in t or
        "swelling" in t or
        "irritation" in t or
        "dermatitis" in t or
        "hives" in t or
        "urticaria" in t or
        "edema" in t
    ):
        return "Inflammation/Skin Reactions"
    
    # ---------- NEUROLOGICAL/SENSORY SYMPTOMS ----------
    elif (
        "blindness" in t or
        "low vision" in t or
        "deafness" in t or
        "hearing loss" in t or
        "dizziness" in t or
        "convulsions" in t or
        "seizures" in t or
        "loss of consciousness" in t or
        "numbness" in t or
        "tingling" in t or
        "paralysis" in t or
        "paraplegia" in t or
        "quadriplegia" in t
    ):
        return "Neurological/Sensory Symptoms"

    # ---------- PSYCHOLOGICAL/ANXIETY/STRESS ----------
    elif (
        "anxiety" in t or
        "stress" in t or
        "panic disorder" in t
    ):
        return "Psychological"

    # ---------- PAIN/SORENESS/HURT ----------
    elif (
        "pain" in t or
        "soreness" in t or
        "hurt" in t or
        "headache" in t or
        "chest pain" in t or
        "abdominal pain" in t
    ):
        return "Pain"

    # -----------------------------------------------------------
    # 6. NON-INJURY/GENERAL/UNSPECIFIED
    # -----------------------------------------------------------
    
    # ---------- DAMAGE TO MEDICAL DEVICES/OTHER DEVICES ----------
    elif (
        "damage to prosthetic" in t or
        "damage to medical implants" in t or
        "damage to artificial limb" in t or
        "damage to replaced joints" in t or
        "damage to other devices" in t
    ):
        return "Damage To Medical Devices"

    # ---------- DENTAL INJURIES ----------
    elif "dental injuries" in t:
        return "Dental Injuries"

    # ---------- MULTIPLE/COMPLEX/GENERAL TRAUMA/UNSPECIFIED INJURY ----------
    # This acts as a final specific and non-specific trauma catch-all.
    elif (
        "multiple" in t or
        "traumatic" in t or
        "trauma" in t or
        "nonclassifiable" in t or
        "nonspecified" in t or
        "n.e.c." in t or
        "n.e.c" in t or
        "unspecified" in t
    ):
        return "Multiple/Unspecified Trauma"

    # ---------- SYMPTOMS/SIGNS/ILL-DEFINED CONDITIONS ----------
    elif (
        "symptom" in t or
        "signs" in t or
        "malaise" in t or
        "fatigue" in t or
        "nausea" in t or
        "vomiting" in t or
        "ill-defined conditions" in t
    ):
        return "General Symptoms"
        
    # Final default catch-all
    else:
        return "Unspecified"
```

```python
#Applying the function
OSHA1["Injury Type"] = OSHA1["NatureTitle"].apply(classify_nature_type)
```

```python
#Creating a function to categorize events
def add_perfect_event_category(OSHA1):
    
    df = OSHA1.copy()
    t = df['EventTitle'].fillna('').astype(str).str.lower().str.strip()

    conditions = [
        # 1. Falls  Same Level
        t.str.contains('fall on same level|slip.*without fall|trip.*without fall|fall while sitting|fall onto or against object'),

        # 2. Falls  To Lower Level
        t.str.contains('fall to lower level|fall through surface|fall from collapsing|jump to lower level|curtailed by personal fall arrest'),

        # 3. Struck By Falling/Flying/Rolling/Dropped Object
        t.str.contains('struck by.*(falling|dropped|flying|discharged|rolling|dislodged|swinging|slipping|load|tipping|part still attached)'),

        # 4. Struck By  General / Unspecified
        t.str.contains('struck by') & ~t.str.contains('falling|dropped|flying|discharged|rolling|dislodged|swinging|slipping|load|tipping'),

        # 5. Struck Against / Bumped Into
        t.str.contains('struck against|walked.*into|ran into|stepped on object'),

        # 6. Caught In / Between / Compressed / Pinched (NO collapse crushing!)
        t.str.contains('caught in|caught between|compressed|pinched|entangled|part of occupant.*caught'),

        # 7. Overexertion & Bodily Reaction
        t.str.contains('overexertion|bodily reaction|bodily conditions|bending|twisting|reaching|crawling|kneeling|running.*without|walking.*without|climbing or stepping'),

        # 8. Contact with Heat / Cold
        t.str.contains('hot objects|cold objects|environmental heat|environmental cold|ignition of clothing'),

        # 9. Exposure to Electricity
        t.str.contains('electricity|electric arc'),

        # 10. Exposure to Harmful Substances
        t.str.contains('inhalation|ingestion|exposure through|needlestick|harmful substance|harmful environment') & ~t.str.contains('electricity'),

        # 11. Fire or Explosion
        t.str.contains('fire|explosion|ignition|flash|dust explosion|blast'),

        # 12. Vehicle Incidents (collision, overturn, ran off, etc.)
        t.str.contains('collision|jack-knifed|overturned|ran off|derailment|sudden start or stop|vehicle struck object'),

        # 13. Pedestrian Struck by Vehicle
        t.str.contains('pedestrian struck|pedestrian vehicular'),

        #  # 14. Assault / Violence / Restraining
        t.str.contains('hitting|kicking|beating|shooting|stabbing|strangulation|restraining|subduing|horseplay|violence|injured by physical contact'),

        # 15. Animal / Insect
        t.str.contains('animal|bitten|stung|kicked by animal|gored|rammed|trampled|mauled|clawed|insect'),

        # 16. Cut / Puncture by Hand Tool or Object
        t.str.contains('injured by.*handheld|slipping or swinging object held|object breaking in hand|cut|stab|slash|pierce'),

        # 17. Rubbed / Abraded
        t.str.contains('rubbed|abraded|friction|foreign matter in eye|kneeling on surface'),

        # 18. Collapse / Cave-in / Crushed by Structure or Material
        t.str.contains('collapsing structure|cave-in|engulfment|excavation|trenching|crushed in collapsing|struck.*caught.*crushed'),

        # 19. Contact with Objects & Equipment  n.e.c. / Unspecified
        t.str.contains('contact with objects|contact.*n.e.c.|struck.*n.e.c.|caught.*n.e.c.|contact incidents'),

        # 20. Nonclassifiable / Truly Unspecified
        t.str.contains('nonclassifiable|event or exposure unspecified|n.e.c.$|unspecified$|^$')
    ]

    categories = [
        "Fall - Same Level",
        "Fall - To Lower Level",
        "Struck By Falling/Flying Object",
        "Struck By Object (General)",
        "Struck Against Object",
        "Caught In/Between",
        "Overexertion & Bodily Reaction",
        "Contact with Heat/Cold",
        "Exposure to Electricity",
        "Exposure to Harmful Substances",
        "Fire or Explosion",
        "Vehicle Incident",
        "Pedestrian Struck by Vehicle",
        "Assault or Violence",
        "Animal or Insect Injury",
        "Cut or Puncture by Tool",
        "Friction or Abrasion",
        "Collapse or Crushed by Structure",
        "Contact with Objects n.e.c.",
        "Nonclassifiable"
    ]

    df['EventCategory'] = np.select(conditions, categories, default="Other")

    # Final cleanup
    df['EventCategory'] = df['EventCategory'].replace('', 'Other')

    return df

# JUST RUN THIS ONE LINE:
OSHA1 = add_perfect_event_category(OSHA1)
```
